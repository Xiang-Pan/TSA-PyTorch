# split

```python
target_dict=dict()
for i in range(len(alldata)):
    target=alldata[i]['text_aspect']
    target=stemmer.stemWord(target)
    if target not in target_dict:
        l=list()
        l.append(i)
        target_dict[target]=l
    else:
        l=target_dict[target]
        l.append(i)
        target_dict[target]=l
        

        
    
```

```python
len(target_dict)
```

```python
target_dict
```

```python
train_val_dict = {key: value for i, (key, value) in enumerate(target_dict.items()) if i % 5 != 0}
test_dict = {key: value for i, (key, value) in enumerate(target_dict.items()) if i % 5 == 0}
```

```python
len(train_val_dict)
```

```python
len(test_dict)
```

```python
train_dict = {key: value for i, (key, value) in enumerate(train_val_dict.items()) if i % 20 != 0}
val_dict = {key: value for i, (key, value) in enumerate(train_val_dict.items()) if i % 20 == 0}
```

```python
len(train_dict)
```

```python
train_index=[]
for key,value in train_dict.items():
    train_index.extend(value)
train_index
```

```python
test_index=[]
for key,value in test_dict.items():
    test_index.extend(value)
test_index
```

```python
val_index=[]
for key,value in val_dict.items():
    val_index.extend(value)
val_index
```
