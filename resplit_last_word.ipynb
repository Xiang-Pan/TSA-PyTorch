```python
target_dict=dict()
for i in range(len(alldata)):
    target=alldata[i]['text_aspect'].strip().split()[-1]
    if target not in target_dict:
        l=list()
        l.append(i)
        target_dict[target]=l
    else:
        l=target_dict[target]
        l.append(i)
        target_dict[target]=l
        

        
    
```
